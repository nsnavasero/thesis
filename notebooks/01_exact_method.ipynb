# Exact Method

This notebook runs the exact single-site dissipation solution and saves the result to a CSV file.

Current assumptions:
- Fock state mean: $N(t)=N_0 e^{-\gamma t}$
- Fock state variance: $N_0(e^{-\gamma t} - e^{-2\gamma t})$
- Coherent state variance: equal to the mean under pure loss

Example filename:
- `exact_fock_numOfParticles5_gamma1p0_time5p0_dt0p001_numOfSamples1.csv`


```python
from pathlib import Path
import sys

PROJECT_ROOT = Path.cwd().resolve().parent
SRC_DIR = PROJECT_ROOT / "src"
if str(SRC_DIR) not in sys.path:
    sys.path.append(str(SRC_DIR))

OUTPUT_DIR = PROJECT_ROOT / "outputs"
OUTPUT_DIR.mkdir(exist_ok=True)
```

```python
import matplotlib.pyplot as plt

from bosonic_dissipation.exact_method import run_exact_and_save

global_defaults = {
    "initial_state_type": "fock",
    "num_of_particles": 5,
    "gamma": 1.0,
    "time": 5.0,
    "dt": 1e-3,
    "num_of_samples": 1,
    "prefer_gpu": True,
}

exact_config = {
    "initial_state_type": global_defaults["initial_state_type"],
    "num_of_particles": global_defaults["num_of_particles"],
    "gamma": global_defaults["gamma"],
    "time": global_defaults["time"],
    "dt": global_defaults["dt"],
    "num_of_samples": 1,
    "prefer_gpu": global_defaults["prefer_gpu"],
}

exact_config
```

## Runtime Block: Exact

This block is intentionally isolated so each numerical method can follow the same pattern later.


```python
exact_result, exact_csv_path = run_exact_and_save(
    OUTPUT_DIR,
    initial_state_type=exact_config["initial_state_type"],
    num_of_particles=exact_config["num_of_particles"],
    gamma=exact_config["gamma"],
    time=exact_config["time"],
    dt=exact_config["dt"],
    num_of_samples=exact_config["num_of_samples"],
    prefer_gpu=exact_config["prefer_gpu"],
)

print(f"Backend used: {exact_result.backend}")
print(f"Runtime (s): {exact_result.runtime_seconds:.6f}")
print(f"Saved CSV: {exact_csv_path}")
```

```python
fig, axes = plt.subplots(1, 2, figsize=(12, 4))

axes[0].plot(exact_result.time_values, exact_result.mean_particle_number, label="Exact")
axes[0].set_title("Mean Particle Number")
axes[0].set_xlabel("Time")
axes[0].set_ylabel("Mean Particle Number")
axes[0].grid(True, alpha=0.3)
axes[0].legend()

axes[1].plot(exact_result.time_values, exact_result.variance, label="Exact", color="tab:orange")
axes[1].set_title("Variance")
axes[1].set_xlabel("Time")
axes[1].set_ylabel("Variance")
axes[1].grid(True, alpha=0.3)
axes[1].legend()

plt.tight_layout()
plt.show()
```
